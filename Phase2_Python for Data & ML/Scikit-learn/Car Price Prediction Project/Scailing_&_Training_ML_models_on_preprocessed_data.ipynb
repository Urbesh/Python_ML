First we will scale the continuous numeric vales

```python
import pandas as pd
```

```python
data=pd.read_csv("dataset\Training_set_processed.csv")
```

```python
data.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
num_cols = ['Levy', 'Engine volume', 'Mileage(Km)', 'Cylinders','Airbags','Age']
scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
data['Price_log'] = np.log1p(data['Price'])
print(data[num_cols].describe())
```

Data is Scaled and now we are ready to train and test the data in multiple ML models 

```python
X=data.drop(columns=['Price', 'Price_log'])  # all independent variables
Y=data['Price_log'] # dependent variable (log of price)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
# Decision Tree
tree_reg=DecisionTreeRegressor(random_state=42)
tree_reg.fit(X,Y)
# Random Forest
forest_reg=RandomForestRegressor(random_state=42)
forest_reg.fit(X,Y)
# Predict using training data
lin_preds=lin_reg.predict(X)
tree_preds=tree_reg.predict(X)
forest_preds=forest_reg.predict(X)
# Calculate RMSE
lin_rmse=np.sqrt(mean_squared_error(Y, lin_preds))
tree_rmse=np.sqrt(mean_squared_error(Y, tree_preds))
forest_rmse=np.sqrt(mean_squared_error(Y, forest_preds))

print("Linear Regression RMSE:", lin_rmse)
print("Decision Tree RMSE:", tree_rmse)
print("Random Forest RMSE:", forest_rmse)
```

Cross-Validation on Decision Tree

```python
from sklearn.model_selection import cross_val_score
import pandas as pd
tree_rmses = -cross_val_score(
    tree_reg,
    X,
    Y,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print("Decision Tree CV RMSEs:", tree_rmses)
print("\nCross-Validation Performance (Decision Tree):")
print(pd.Series(tree_rmses).describe())
```

Cross-Validation on Linear Regression

```python
lin_rmses = -cross_val_score(
    lin_reg,
    X,
    Y,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print("Linear Regression CV RMSEs:", lin_rmses)
print("\nCross-Validation Performance (Linear Regression):")
print(pd.Series(lin_rmses).describe())
```

Cross-Validation on Random Forest

```python
forest_rmses = -cross_val_score(
    forest_reg,
    X,
    Y,
    scoring="neg_root_mean_squared_error",
    cv=10
)
print("Random Forest CV RMSEs:", forest_rmses)
print("\nCross-Validation Performance (Random Forest):")
print(pd.Series(forest_rmses).describe())
```
