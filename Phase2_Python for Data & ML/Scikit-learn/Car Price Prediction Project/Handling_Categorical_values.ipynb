Computer doesn't understands words it understands numbers, So we need to convert the categorical values to numbers that or Machine Learning model will understand.


Since this dataset have unordered categories we wil use One-Hot encoding.


```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

Let's start with the manufacturer column

```python
man=data[["Manufacturer"]]
```

```python
man_cat=encoder.fit_transform(man)
```

```python
man_cat.toarray()
```

```python
encoder.categories_
```

```python
man_cat=pd.DataFrame(man_cat.toarray(), columns=encoder.categories_[0],index=man.index)
man_cat.columns = ['Manufacturer_' + col for col in man_cat.columns]
man_cat.head(10)
```
