```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv("dataset/Iris_train.csv")
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
model=RandomForestClassifier()
model.fit(x,y)
```

```python
data_test=pd.read_csv("dataset/Iris_test.csv")
```

```python
X_test=data_test.iloc[:,:-1]
predictions=model.predict(X_test.values)
```

```python
data_test["Predictions"]=predictions
data_test
```
