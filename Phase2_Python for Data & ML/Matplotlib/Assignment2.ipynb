**Assignment:**  
Create a side-by-side bar chart comparing runs of Dhoni, Kohli, and Rohit over 10 selected years.  
Then create a horizontal bar chart showing total runs scored in the first 5 years.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
Rohit=[504,611,168,1196,578,815,564,1293,1030,1490]
Dhoni=[600,764,524,753,418,640,278,788,275,600]
Virat=[995,1381,1026,1268,1054,623,739,1460,1202,1377]
```

```python
x=np.arange(len(year))
width=0.25
plt.style.use("seaborn-v0_8-bright")
plt.bar(x-width,Virat,width=width,label="Virst")
plt.bar(x,Dhoni,width=width,label="Dhoni")
plt.bar(x+width,Rohit,width=width,label="Rohit")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Bar Comparison Graph")
plt.xticks(x,year)
plt.grid(True)
plt.legend()
```

```python
players=["Rohit","Dhoni","Virat"]
Run_fiveyear=[504+611+168+1196+578, 600+764+524+753+418, 995+1381+1026+1268+1054]
plt.style.use("seaborn-v0_8-bright")
plt.barh(players,Run_fiveyear)
plt.xlabel("Total Runs Scored During First Five Years (2010-2014)")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.grid(True)
```
