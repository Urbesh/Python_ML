**Assignment:**  
Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.

---

```python
import matplotlib.pyplot as plt
```

```python
OS=["Android", "iOS", "SamsungOS", "Others"]
explode=[0,0.1,0,0]
shares=[72.15,22.2,2.46,3.19]
plt.style.use("seaborn-v0_8-bright")
plt.pie(shares,labels=OS,explode=explode,shadow=True,autopct='%1.1f%%')
plt.title("Mobile OS Market Share")
plt.legend(loc="upper center")
```

```python
plt.barh(OS,shares,color='cyan')
plt.xlabel("Shares in percentage(%)")
plt.ylabel("Operating Systems")
plt.title("Mobile OS Market Share")
plt.grid(True)
```
