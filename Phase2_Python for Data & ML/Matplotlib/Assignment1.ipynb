Create a plot comparing **Kohli**, **Rohit Sharma**, and **Sehwag** across 10 years of hypothetical runs.  
Use:
- Labels

```python
import matplotlib.pyplot as plt
```

```python
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
Virat=[1644,2186,1913,2286,1307,2595,2818,2735,2455,842]
Rohit=[611,168,1196,578,815,564,1293,1030,1490,171]
Sehwag=[1294, 983, 43,0, 0, 0, 0, 0, 0, 0]
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.plot(year, Virat, color='orange', linestyle='--', label="Kohli")
plt.plot(year, Rohit, color='green', linestyle='-.', label="Rohit")
plt.plot(year, Sehwag, color='blue', label="Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.grid(True)
plt.legend()
```
