Try making a stackplot with your own weekly schedule!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("My_routine.csv")
df
```

```python
days=df["Day"].to_numpy()
gym=df["GYM(hrs)"].to_numpy()
collage=df["Collage(hrs)"].to_numpy()
self=df["Self Study(hrs)"].to_numpy()
discord=df["Discord(hrs)"].to_numpy()
sleep=df["Sleep(hrs)"].to_numpy()
misc=df["Miscellaneous(hrs)"].to_numpy()
labels=["GYM","Collage","Self Study","Discord","Sleep","Miscellaneous"]
```

```python
plt.figure(figsize=(10,6))
plt.style.use("seaborn-v0_8-bright")
plt.stackplot(days,gym,collage,self,discord,sleep,misc,labels=labels)
plt.legend(loc="upper left")
plt.title("Weekly Activity Tracker")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.grid(True)
```
