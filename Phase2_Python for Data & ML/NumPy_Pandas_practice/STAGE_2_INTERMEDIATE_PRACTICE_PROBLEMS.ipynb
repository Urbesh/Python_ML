```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
df1=pd.read_csv("city_population.csv")
df1
```

```python
df2=pd.merge(df,df1,on="City")
df2
```

```python

df2['Revenue'] = df2['Price'] * df2['Quantity']
df2
```

```python
df2.groupby('City')['Revenue'].sum()
```

```python
pivot_rev = pd.pivot_table(df2, values='Revenue', index='Category', columns='City', aggfunc='sum', fill_value=0, margins=True)
pivot_rev
```
