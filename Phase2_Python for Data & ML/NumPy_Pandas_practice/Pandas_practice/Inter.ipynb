```python
import pandas as pd
```

```python
df_given=pd.read_csv("employees.csv")
df_given.info()
```

```python
df_given.isnull()
```

```python
#Replace missing values with the mean salary.
df_given["Salary"]=df_given["Salary"].fillna(df_given["Salary"].mean())
```

```python
df_given.info()
```

```python
#Then drop any remaining rows with missing data.
df_given=df_given.dropna()
```

```python
df_given.info()
```

```python
#Create a DataFrame of employees and sort them:
    # Ascending by Department
df_given.sort_values("DeptID")
```

```python
#Create a DataFrame of employees and sort them:   
    #Descending by Salary
df_given.sort_values("Salary", ascending=False)
```

```python
#Rename column Name to Employee_Name and reset the index
df_given.rename(columns={"Name":"Employee_Name"}, inplace=True)
```

```python
#Given a DataFrame of student marks in subjects, use melt() to convert it into long format
df_marks=pd.read_csv("marks.csv")
df_marks
```

```python
df_marks.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name="Subject",value_name="Score")
```
